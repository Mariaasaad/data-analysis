```python
#1.	Make a chart to illustrate if weekly sales are increasing or decreasing over time.
sns.lineplot( x= 'Date', y= 'Weekly_Sales', data= df)

plt.xlabel('date')
plt.ylabel('Weekly_Sales')
plt.title('weekly sales over time')

plt.show()
```

```python
#2.	Make a chart to show how much each brand sells.

store_by_Weekly_Sales = df.groupby('Store')['Weekly_Sales'].sum().reset_index()
display(store_by_Weekly_Sales)
```

```python
store_by_Weekly_Sales.plot.bar( )

plt.xlabel('Store')
plt.ylabel('Weekly_Sales')
plt.title('show how much each brand sells')
```

```python
#3.	Determine the top ten selling stores.

store_by_Weekly_Sales = df.groupby('Store')['Weekly_Sales'].sum().reset_index()
display(store_by_Weekly_Sales)
```

```python
store_by_Weekly_Sales = store_by_Weekly_Sales.sort_values('Weekly_Sales',ascending=False)[:10]
display(store_by_Weekly_Sales)
```

```python
top_ten_stores = store_by_Weekly_Sales['Store'].values
display(top_ten_stores)
```

```python
#4.	Make a histogram to show the top 10 stores sales.
sns.barplot(x='Store' , y='Weekly_Sales' , data=store_by_Weekly_Sales , color= 'red')

plt.xlabel('top_ten_stores')
plt.ylabel('Weekly_Sales')
plt.title('top ten selling stores')

plt.show
```

```python
#5.	Create a chart that compares average weekly sales for the top ten selling stores during holidays and non-holidays.

store_by_Weekly_Sales = df.groupby('Store')['Weekly_Sales'].sum().reset_index()
display(store_by_Weekly_Sales)
```

```python
store_by_Weekly_Sales = store_by_Weekly_Sales.sort_values('Weekly_Sales',ascending=False)[:10]
display(store_by_Weekly_Sales)
```

```python
top_ten_stores = store_by_Weekly_Sales['Store'].values
display(top_ten_stores)
```

```python
top_10_sales = df[df['Store'].isin(top_ten_stores)]
display(top_10_sales)
```

```python
top_10_stores = df['Store'].value_counts().nlargest(10).index
display(top_10_stores)
```

```python
top_10_sales = df[df['Store'].isin(top_10_stores)]
display(top_10_sales)
```

```python
holiday_avg_sales = top_10_sales[top_10_sales['Holiday'] == 1].groupby('Store')['Weekly_Sales'].mean()
non_holiday_avg_sales = top_10_sales[top_10_sales['Holiday'] == 0].groupby('Store')['Weekly_Sales'].mean()

display(holiday_avg_sales)
display(non_holiday_avg_sales)
```

```python
comparison_df = pd.DataFrame({
    'Holiday Avg. Sales': holiday_avg_sales,
    'Non-Holiday Avg. Sales': non_holiday_avg_sales})

display(comparison_df)
```

```python
comparison_df.plot(kind='bar', figsize=(10, 6))
plt.title('Average Weekly Sales Comparison - Top 10 Stores')
plt.xlabel('Store')
plt.ylabel('Average Weekly Sales')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#6.	Create a chart that displays the average weekly sales for each brand department for the top 10 selling stores.

for i in top_ten_stores:
    average_weekly_sales_department = df[df ['Category'] == i].groupby('Store')['Weekly_Sales'].mean()
  
    sns.barplot(x='Category' ,y='Weekly_Sales', data=df ,ci=None )
    plt.title(f"average_weekly_sales_department {i}")
    plt.show()
```

```python
#7.	Make a line chart to show the relationship between weekly sales and weather Temperature.

sns.lineplot(x= 'Temperature', y= 'Weekly_Sales', data=df )

plt.xlabel('Temperature')
plt.ylabel('Weekly_Sales')
plt.title('relationship between weekly sales and weather Temperature.')


plt.show()
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import preprocessing
```

```python
df = pd.read_csv('bazoukaaa.csv' , header=0)
df
df.isna().sum()
```

```python
x = df.drop(columns = ['Date' , 'Weekly_Sales'])
x
```

```python
y= df ['Weekly_Sales']
y
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
```

```python
lb = preprocessing.LabelEncoder()
x.Holiday= lb.fit_transform(x.Holiday)
display (x)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
x[['Fuel_Price']] = scaler.fit_transform(x[['Fuel_Price']])
x
```

```python
df.isna().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['DayOfWeek'] = df['Date'].dt.dayofweek
display(df)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
model= DecisionTreeRegressor()
model.fit(x_train , y_train)
prediction = model.predict(x_test)
score = r2_score(y_test,prediction)
print("r2_score=", score)
print(print("Decision Tree score: {}%".format(round(score * 100, 2))))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
# Model 2: Random Forest Regressor
from sklearn.metrics import mean_squared_error

model2 = RandomForestRegressor()
model2.fit(x_train, y_train) 

# Predictions from model 2
predictions2 = model2.predict(x_test)
accuracy2 = r2_score(y_test, predictions2)

print("Mean Squared Error of Model 2:", accuracy2)
print("Decision Tree Accuracy: {}%".format(round(accuracy2 * 100, 2)))
```
